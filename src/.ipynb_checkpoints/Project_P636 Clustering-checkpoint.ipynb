### Handling Missing values

```python
countries = df['Country']              # categorical column
df_num = df.drop(columns=['Country'])  # only numeric columns
```

```python
# Remove $ and , from all columns and convert to numeric
for col in df_num.columns:
    df_num[col] = (
        df_num[col]
        .astype(str)
        .str.replace('$', '', regex=False)
        .str.replace(',', '', regex=False)
    )
    df_num[col] = pd.to_numeric(df_num[col], errors='coerce')
```

```python
# dropping country column as it is categorical
```

```python
median_cols = [
    'GDP',
    'CO2emissions',
    'energy usage',
    'healthexp/capita',
    'infant mortality'
]
```

```python
for col in median_cols:
    if col in df_num.columns:
        df_num[col].fillna(df_num[col].median())
```

```python
for col in df_num.columns:
    if col not in median_cols:
        df_num[col]=df_num[col].fillna(df_num[col].mean())
```
